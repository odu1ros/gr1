```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# data_prev = pd.read_excel('data/WIOTS_in_EXCEL/WIOT2013_Now16_ROW.xlsb', engine='pyxlsb')
# data_current = pd.read_excel('data/WIOTS_in_EXCEL/WIOT2014_Nov16_ROW.xlsb', engine='pyxlsb')
data_prev = pd.read_excel('data/wiot_2013.xlsx')
data_current = pd.read_excel('data/wiot_2014.xlsx')
data_current_pyp = pd.read_excel('data/wiot_2014_pyp.xlsx')
data_currency = pd.read_excel('data/Exchange_Rates.xlsx', sheet_name=1, header=3).iloc[36, :]
```

```python
data_prev.iloc[1:, 2:] = data_prev.iloc[1:, 2:].replace(0, np.NaN) / data_currency['_2013']
data_current.iloc[1:, 2:] = data_current.iloc[1:, 2:].replace(0, np.NaN) / data_currency['_2014']
data_current_pyp.iloc[1:, 2:] = data_current_pyp.iloc[1:, 2:].replace(0, np.NaN) / data_currency['_2013']
```

```python
data_current_pyp.head()
```

```python
c0q0 = data_prev.iloc[57, 2:58]
c0q1 = data_current_pyp.iloc[57, 2:58]
c1q1 = data_current.iloc[57, 2:58]

p0q0 = data_prev.iloc[-1, 2:58]
p0q1 = data_current_pyp.iloc[-1, 2:58]
p1q1 = data_current.iloc[-1, 2:58]

c0 = c0q0 / p0q0
c1 = c1q1 / p1q1
c1_p0 = c1 * p1q1 / p0q0
p1_p0 = p1q1 / p0q1

tot_p1q1 = p1q1.sum()
tot_c1q1 = c1q1.sum()
tot_p0q0 = p0q0.sum()

theta_0 = 2 - c0
theta_1 = theta_0 * tot_p1q1 / tot_p0q0
delta_theta = theta_1 / theta_0

p_opt = (theta_1 + c1_p0) / 2
p_coef = p_opt - p1_p0
```

```python
p_coef
```

```python
plt.figure(figsize=(30, 30))
plt.gca().set_ylim([min(p_coef), max(p_coef)])
for i in p_coef.index:
    if p_coef[i] < 0:
        plt.axvline(x=i, ymin=p_coef[i], ymax=0, color='b')
    else:
        plt.axvline(x=i, ymin=0, ymax=p_coef[i], color='b')
# plt.scatter(p_coef.index, p_coef)
plt.show()
```
